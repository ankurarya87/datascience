### Pull Property Sites Details using Zillow API
Define functions needed to parse and extract useful features from results of HTTP requests to Zillow API<p>
More information on Zillow API requests can be found here 
https://www.zillow.com/howto/api/GetRegionChildren.htm

```python
# define function to extract string from parsed json object

def get_str (parent,child):
    #check if there are any children
    if parent.find(child)==None:
        return np.nan
    elif parent.find(child).string==None:
        return np.nan
    else:
        return parent.find(child).string
```

```python
# define function to extract float from parsed json object

def get_float (parent,child):
    #check if there are any children
    if parent.find(child)==None:
        return np.nan
    elif parent.find(child).string==None:
        return np.nan
    elif re.sub('[^\d]' , '', parent.find(child).string)=='':
        return np.nan
    else:
        return float(re.sub('[^\-\d.]' , '', parent.find(child).string))
```

```python
# define function to extract price of property from parsed json object

def get_price (parent,child):
    #check if there are any children
    if parent.find(child,{'currency':['USD']})==None:
        return np.nan
    elif parent.find(child,{'currency':['USD']}).string==None:
        return np.nan
    else:
        return float(parent.find(child,{'currency':['USD']}).string.replace(',',''))
```

```python
# define function to parse useful features like address, geographical coordinates, price, property features

def get_property_details(soup):
        
    # create empty dictionary to store property features from parsed object/ tree 
    dict={'street': [], 'zipcode' : [], 'city' : [], 'state' : [], 'latitude' : [], 'longitude' : [], 
          'FIPScounty' : [], 'useCode' : [], 'lotSizeSqFt' : [], 'finishedSqFt' : [], 
          'bathrooms' : [], 'totalRooms' : [], 'amount' : [], 'link' : [], 'zindex' : []}
    
    # Each response contains multiple property corresponding to street address + zip code HTTP requests
    # Iterate through all properties within the parsed tree 
    for result in soup.response.find_all('result'):
        
        address = result.address
        
        dict['street'].append(get_str(address,'street'))
        dict['zipcode'].append(get_float(address,'zipcode'))
        dict['city'].append(get_str(address,'city'))
        dict['state'].append(get_str(address,'state'))
        dict['latitude'].append(get_float(address,'latitude'))
        dict['longitude'].append(get_float(address,'longitude'))
        dict['FIPScounty'].append(get_float(result, 'FIPScounty'))
        dict['useCode'].append(get_str(result, 'useCode'))
        dict['lotSizeSqFt'].append(get_float(result, 'lotSizeSqFt'))
        dict['finishedSqFt'].append(get_float(result, 'finishedSqFt'))
        dict['bathrooms'].append(get_float(result, 'bathrooms'))
        dict['totalRooms'].append(get_float(result, 'totalRooms'))
        
        zestimate=result.find('zestimate')
        dict['amount'].append(get_price(zestimate,'amount'))
        
        links=result.find('links')
        dict['link'].append(get_str(links,'homedetails'))
        
        localRealEstate = result.find('localRealEstate') 
        dict['zindex'].append(get_float(localRealEstate,'zindexValue'))
        # all property features are extracted into 
        print ("Success! property detail found!")
    
    # If response from zillow api is valid with meaningful property features, 
    # then dataframe is successful created from above dictionary. Otherwise, None is returned.
    try:
        df = pd.DataFrame.from_dict(dict)
        print ("Success! property dataframe submitted")
        return df
    except:
        print ("Failure! property dataFrame not formed")
        return None
    
```

```python
# define function to save the parsed output into xml files for safekeeping in specified folder
def save_xml_file(soup, street, zipcode):
    file_name = street + '_' + str(zipcode)[0:5] + '.xml'
    with open(os.path.join('zillow_data',file_name), "w+") as file:
        file.write(str(soup))
```

```python
# create function which sends HTTP requests to Zillow API by using only street address and  zipcode from open database
def zillow_connect(street, zipcode, city = '', state ='NY'):
    zws_id = 'X1-ZWz1814wcv2kuj_517ck'
    citystatezip = city + ',' + state + ',' + str(zipcode)[0:5]
    url='http://www.zillow.com/webservice/GetDeepSearchResults.htm?zws-id='+zws_id+'&address='+ street \
    + '&citystatezip=' + citystatezip
    print("Request :\n Street = %s, Zip Code =%s" %(street,zipcode))
    num_of_attempt = 0
    while True:
        try:
            results = requests.get(url)
            print ("Connection Successful!")
            return results
            break
        except:
            print ("Failure to Connect ...Try Again")
```

For a successful property detail query using Zillow API, atleast street address and zipcode are required. 
Two open sources were found useful <p>
(1) Open address data from openaddress.io [csv file]
 https://s3.amazonaws.com/data.openaddresses.io/runs/600731/us/ny/city_of_new_york.zip
 
(2) Open street address from NYC open data [shape file]
https://data.cityofnewyork.us/City-Government/NYC-Address-Points/g6pj-hd8k

We are using (1) as an example for this code 

```python
# import file with address of NYC homes. Only street and zipcode are good enough to use with zillow API,
# because it returns results from all properties on the street & zipcode
nyc_openaddress = pd.read_csv('city_of_new_york.csv')
street_array = nyc_openaddress['STREET'].values
zipcode_array = nyc_openaddress['POSTCODE'].values
nyc_openaddress.head()
```

```python
# Initialize a dataframe to store all property feature values 
results_df = pd.DataFrame(columns={'street', 'zipcode', 'city', 'state', 'latitude', 'longitude', 
          'FIPScounty', 'useCode', 'lotSizeSqFt', 'finishedSqFt', 
          'bathrooms', 'totalRooms', 'amount', 'link', 'zindex'})

# Here is the main code, which brings all the functions defined above into action 
# Iterate through list of street + zipcodes from open address of NYC
for street, zipcode in zip(street_array, zipcode_array):

    # gather results by sending HTTP request to zillow 
    results = zillow_connect(street, zipcode)

    # parse the XML response into a tree structrure
    soup = BeautifulSoup(results.text, "xml")   

    # save the parsed reponse in a xml file as backup
    save_xml_file(soup, street, zipcode)
  
    #check if the response is valid
    if soup.find('response')!=None:
        if soup.response.find('result')!=None:
            print ("Valid Response!, Begin Extraction ...")
            # extract useful property features
            newresult_df = get_property_details(soup)
        else:
            print ("Invalid Response!, Next Iteration ...")
            continue
    else:
        print ("Invalid Response!, Next Iteration ...")
        continue
        
    # combine results into 
    results_df = pd.concat([results_df, newresult_df], join='inner', sort=False) 
    results_df.reset_index(drop=True,inplace=True)
    print("------------------------------------")

# End the finish  
print("-------------FINISHED-----------------")
```

```python
# Check the aggregated results 
print ("No. of sites collected from zillow API :", results_df.shape[0])
results_df.head()
```

```python
# save the results in csv file
results_df.to_csv('zillow_out.csv')
```

### Determine Residential Property Price Distrubution by NYC Boroughs

```python
# Create subplots
fig, axes = plt.subplots(1,5, sharex=True, sharey=True, figsize=(16, 5))

# Define variables to be used in plots
boroughs = df['borough_name'].unique()
colors=['red','blue','green','orange','purple']

# Iterate through each borough 
for ax, borough, color in zip(axes, boroughs,colors):
    data = df.loc[df['borough_name']== borough]['amount']
    bins = np.logspace(np.log10(1e5), np.log10(1e7), 30)
    ax.hist(data, bins = bins, color = color, density=False)
    ax.set_title(borough)
    ax.set_xscale('log')
    ax.set_ylabel('')
    ax.set_xlabel('Price in USD ($)')
    median_label = 'median\n'+'$'+str(int(data.median()))
    ax.annotate(median_label, xy = (data.median()+1e5,6000), fontsize=14)
    ax.axvline(x=data.median(),linestyle='dotted',lw=2, color = 'black')
axes[0].set_ylabel('Frequency')
fig.suptitle('Residential Property Price Distibution by NYC Boroughs',fontsize=16)
fig.text(x=0.75,y=0.0, s='Ankur Arya 2019 | Data Courtesy : Zillow')
```

### Find Expensive Neighborhoods by NYC Boroughs

```python
# Define subplots 
fig, axes = plt.subplots(1,5, sharex=True, sharey=True,figsize=(20, 5))

# Create variable used in plotting
boroughs = df['borough_name'].unique()
colors=['red','blue','green','orange','purple']

# Iterate through each borough
for ax, borough, color in zip(axes, boroughs,colors):
    data = df.loc[df['borough_name']== borough][['amount','hood_name']]
    data_hood = data.groupby ('hood_name').median()
    data_hood.sort_values(by='amount', ascending=False, inplace=True)
    data_hood.reset_index(drop=False, inplace=True)
    data_plot = data_hood.head()

    ax.barh(y = data_plot.index, width = data_plot['amount'], color = color)
    ax.invert_yaxis()
    ax.set_title(borough)
    ax.set_xscale('log')
    ax.set_ylabel('')
    ax.set_yticks([])
    ax.set_xlabel('Price in USD ($)')
    ax.set_xlim(8e4, 5e7)
    for index,rows in data_plot.iterrows():
        name = rows['hood_name']
        x = rows['amount']
        price = '$'+ str(int(x))
        namesplit = name.split(' ')
        if len(namesplit)>1:
            name1=namesplit[0]
            name2=' '.join(namesplit[1:])
            ax.text(s=name1, x=1e5, y=index-0.2, color='white',ha='left', fontweight='bold',wrap=True)
            ax.text(s=name2, x=1e5, y=index+0.2, color='white',ha='left', fontweight='bold',wrap=True)
        elif (len(name)>8) & (int(x)<680000):
            ax.text(s=name, x=1e5, y=index, color='white',ha='left', fontweight='bold',wrap=True, fontsize=8)
        else:
            ax.text(s=name, x=1e5, y=index, color='white',ha='left', fontweight='bold')
        ax.annotate('  ' + price, xy=(x,index), fontsize=12)
axes[0].set_ylabel('Neighborhoods with median price')
fig.suptitle('Top 5 Most Expensive Neighborhoods for Residential Properties by NYC Boroughs\n',fontsize=16)
fig.text(x=0.75,y=0.0, s='Ankur Arya 2019 | Data Courtesy : Zillow, NYC OpenData')
#plt.tight_layout()
```

### Find commute time to focal points of work & travel

Midtown Manhattan is largest central business district in the world and central portion of borough of Manhattan (https://en.wikipedia.org/wiki/Midtown_Manhattan)

John F. Kennedy International (JFK) Airport is the primary internation airport serving new york city and busiest international air passenger gateway into North America (https://en.wikipedia.org/wiki/John_F._Kennedy_International_Airport)

Let us assume Midtown Manhattan as a focal point if your working in Manhattan and you need to fly frequently from JFK airport. Then the choice of neighborhood should include time to commute to these places. Use Google Distance Matrix API to get time to commute with by car or transit like subway / bus to Midtown Manhattan and JFK airport. You can learn more about google's distance matric API here https://developers.google.com/maps/documentation/distance-matrix/start

```python
# Destination # 1 : Midtown Manhattan
df_destn = df_home.loc[df_home['hood_name'] =='Midtown South'].values.tolist()[0]
midtown_manhattan = df_destn[3:5]
midtown_manhattan

# Destination # 2 : JFK Airport
##Get coordinates from geolocator for maps
address = 'John F. Kennedy International Airport, Queens, NY 11430'
geolocator = Nominatim(user_agent="ny_explorer")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
jfk_airport = [latitude, longitude]
jfk_airport
```

```python
g_api_key = 'AIzaSyBjAxAuw8OQ2hSGZ22IQyYFevbdr8mlbBI'
```

```python
mode = ['driving', 'transit&transit_mode=subway|train|bus']
```

```python
Manhattan_drive_time = []
Manhattan_subway_time = []
JFK_airport_drive_time = []

for index, row in df_home.iterrows():
    origin = [row[3],row[4]]
   
    # Time to drive to Midtown Manhattan
    url = 'https://maps.googleapis.com/maps/api/distancematrix/json?origins={},{}&destinations={},{}&mode={}&key={}'\
    .format(origin[0], origin[1], midtown_manhattan[0], midtown_manhattan[1], mode[0], g_api_key)
    while True:
        try:
            results = requests.get(url)
            print ("Connection Successful!")
            break
        except:
            print ("Failure ! Error code {} ... Trying Again".format(results.status_code))
    result = json.loads(results.text)
    if result['rows'][0]['elements'][0]['status'] == 'OK':
        time_min = result['rows'][0]['elements'][0]['duration']['value'] / 60
    else :
        time_min = np.nan
    print ('Time to drive from {} to {} takes {} mins' .format(row[0], 'Midtown Manhattan', time_min))
    Manhattan_drive_time.append(time_min)
    
    # Time to take transit - subway / bus to Midtown Manhattan
    url = 'https://maps.googleapis.com/maps/api/distancematrix/json?origins={},{}&destinations={},{}&mode={}&key={}'\
    .format(origin[0], origin[1], midtown_manhattan[0], midtown_manhattan[1], mode[1], g_api_key)
    while True:
        try:
            results = requests.get(url)
            print ("Connection Successful!")
            break
        except:
            print ("Failure ! Error code {} ... Trying Again".format(results.status_code))
    result = json.loads(results.text)
    if result['rows'][0]['elements'][0]['status'] == 'OK':
        time_min = result['rows'][0]['elements'][0]['duration']['value'] / 60
    else :
        time_min = np.nan
    print ('Time to take subway from {} to {} takes {} mins' .format(row[0], 'Midtown Manhattan', time_min))
    Manhattan_subway_time.append(time_min)
    
    # Time to drive to JFK airport
    url = 'https://maps.googleapis.com/maps/api/distancematrix/json?origins={},{}&destinations={},{}&mode={}&key={}'\
    .format(origin[0], origin[1], jfk_airport[0], jfk_airport[1], mode[0], g_api_key)
    while True:
        try:
            results = requests.get(url)
            print ("Connection Successful!")
            break
        except:
            print ("Failure ! Error code {} ... Trying Again".format(results.status_code))
    result = json.loads(results.text)
    if result['rows'][0]['elements'][0]['status'] == 'OK':
        time_min = result['rows'][0]['elements'][0]['duration']['value'] / 60
    else :
        time_min = np.nan
    print ('Time to drive from {} to {} takes {} mins' .format(row[0], 'JFK Airport', time_min))
    JFK_airport_drive_time.append(time_min)
```

#### Search restaurants venues near neighborhoods
Use search query 'restaurant' and remove some venue categories like Food, Restaurant, ... which do not represent cuisine. 

```python
# Create empty dataframes to load itemized result and its summary
df_restaurant = pd.DataFrame(columns = {'Neighborhood', 'Venue_Name', 'Venue_Latitude', 'Venue_Longitude', 'Venue_Category'})
df_res_sum = pd.DataFrame(columns = {'Neighborhood', 'Venue_Name', 'Venue_Category'})

# iterate  over final list of neighborhoods
for index, row in df_final.iterrows():
    latitude = row.Latitude
    longitude = row.Longitude
    
    hood_list = []; name_list = []; lat_list = []; lon_list = []; cat_list = [] 
    search_query = 'Restaurant'
    
    # foursquare HTTP request
    url = 'https://api.foursquare.com/v2/venues/search?client_id={}&client_secret={}&ll={},{}&v={}&query={}&radius={}&limit={}'.format(CLIENT_ID, CLIENT_SECRET, latitude, longitude, VERSION, search_query, radius, LIMIT)
    results = requests.get(url).json()
    
    for j, venue in enumerate(results['response']['venues']):
        try:
            name = venue['name']
        except:
            name = np.nan
        
        try:
            lat = venue['location']['lat']
        except:
            lat = np.nan
            
        try:
            lon = venue['location']['lng']
        except:
            lon = np.nan
        
        try:
            cat = venue['categories'][0]['shortName']
        except:
            cat = np.nan
        
        name_list.append(name)
        cat_list.append(cat)
        lat_list.append(lat)
        lon_list.append(lon)
        hood_list.append(row['Neighborhood'])

    df_row = pd.DataFrame({'Neighborhood':hood_list, 'Venue_Name':name_list, \
                    'Venue_Latitude':lat_list, 'Venue_Longitude':lon_list, 'Venue_Category':cat_list})
    
    df_row_sum = df_row.groupby(['Neighborhood','Venue_Category'], as_index = False).agg({'Venue_Name':'count'})
    # filter for undesired venue catergory in the summary dataframe
    df_row_sum = df_row_sum[(df_row_sum['Venue_Category']!='Food') & (df_row_sum['Venue_Category']!='Restaurant')]
    df_row_sum.sort_values(by=['Venue_Name'], ascending = False, inplace = True)
    df_row_sum = df_row_sum.head(5)
    
    df_restaurant = pd.concat([df_restaurant,df_row])
    df_res_sum = pd.concat([df_res_sum, df_row_sum])

df_restaurant.reset_index(drop=True, inplace=True)
df_res_sum.reset_index(drop=True, inplace=True)
```

```python
# Create function to return same color for a venue category. This will be used in tree plots below
def get_color (labels):
    categories = df_res_sum['Venue_Category'].unique().tolist()
    color_list = []
    for label in labels :
        index = categories.index(label)
        if index < 9:
            color = plt.cm.Set1(index)
        else :
            color = plt.cm.Set3(index%12)
        color_list.append(color)
    return color_list

get_color(['Pizza', 'Asian'])
```

```python
# Create Tree plot for each neighborhood to visualize restaurants cuisine / category

fig, axes = plt.subplots(2, len(df_res_sum['Neighborhood'].unique())//2, figsize = (16,8))
nrow , ncol = 0, 0
for index, neighborhood in enumerate(df_res_sum['Neighborhood'].unique().tolist()):
    df_subset = df_res_sum[df_res_sum['Neighborhood'] == neighborhood]
    df_subset['Venue_Name'].astype(int)
    sizes = df_subset['Venue_Name'].apply(lambda y: int(100*y/df_subset['Venue_Name'].sum())).tolist()
    labels = df_subset['Venue_Category'].tolist()
    # get the color for the categories
    colors = get_color(labels)
    # add numeric % for each category label 
    for i, label in enumerate(labels):
        labels[i] = labels[i] + " "+ str(sizes[i]) + "%" 
    
    axes[nrow][ncol].set_axis_off()
    axes[nrow][ncol].set_title(neighborhood, fontsize = 16)
    
    squarify.plot(sizes = sizes, label = labels, color = colors, ax = axes[nrow][ncol], \
                  text_kwargs={'fontsize':14})

    if ncol < (len(df_res_sum['Neighborhood'].unique())//2):
        ncol +=1
    if ncol == (len(df_res_sum['Neighborhood'].unique())//2):
        ncol = 0
        nrow = 1
fig.tight_layout()
fig.text(x=0.70,y=0, s='Ankur Arya 2019 | Data Courtesy : FOURSQUARE, NYC OpenData')
```

#### Search Colleges/University Venues near Neighborhood
searh using query  'College' and filter relavant results using appropriate filters in venue catergory

```python
# Create dataframes to store resutls and its summary by neighborhood
df_college = pd.DataFrame(columns = {'Neighborhood', 'Venue_Name', 'Venue_Latitude', 'Venue_Longitude', 'Venue_Category'})
df_college_sum = pd.DataFrame(columns = {'Neighborhood', 'Venue_Name', 'Venue_Category'})

# iterate through each neighborhood in the final selection list
for index, row in df_final.iterrows():
    latitude = row.Latitude
    longitude = row.Longitude
    
    hood_list = []; name_list = []; lat_list = []; lon_list = []; cat_list = [] 
    search_query = 'College'

    # foursquare API HTTP request
    url = 'https://api.foursquare.com/v2/venues/search?client_id={}&client_secret={}&ll={},{}&v={}&query={}&radius={}&limit={}'.format(CLIENT_ID, CLIENT_SECRET, latitude, longitude, VERSION, search_query, radius, LIMIT)
    results = requests.get(url).json()
    
    for j, venue in enumerate(results['response']['venues']):
        try:
            name = venue['name']
        except:
            name = np.nan
        
        try:
            lat = venue['location']['lat']
        except:
            lat = np.nan
            
        try:
            lon = venue['location']['lng']
        except:
            lon = np.nan
        
        try:
            cat = venue['categories'][0]['shortName']
        except:
            cat = np.nan
        
        name_list.append(name)
        cat_list.append(cat)
        lat_list.append(lat)
        lon_list.append(lon)
        hood_list.append(row['Neighborhood'])

    df_row = pd.DataFrame({'Neighborhood':hood_list, 'Venue_Name':name_list, \
                    'Venue_Latitude':lat_list, 'Venue_Longitude':lon_list, 'Venue_Category':cat_list})
    # filter venue catergory for college / university results
    df_row = df_row[(df_row['Venue_Category'] == 'University')\
                    | (df_row['Venue_Category'] == 'Community College')\
                    | (df_row['Venue_Category'] == 'General College & University')]
    
    df_row_sum = df_row.groupby(['Neighborhood','Venue_Category'], as_index = False).agg({'Venue_Name':'count'})
    df_row_sum = df_row_sum.head(5)
    
    df_college = pd.concat([df_college,df_row])
    df_college_sum = pd.concat([df_college_sum, df_row_sum])

df_college.reset_index(drop=True, inplace=True)
df_college_sum.reset_index(drop=True, inplace=True)
```
